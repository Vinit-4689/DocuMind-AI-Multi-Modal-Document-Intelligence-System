###Extracting Data

####Extration from pdf

```python
import pdfplumber

#funtion
def extract_text_from_pdf(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            text += f"\n--- Page {page_num+1} ---\n"
            text += page.extract_text() or ""
    return text
```

```python
import pytesseract      #libs using for png
from PIL import Image

def extract_text_from_image(image_path):
    image = Image.open(image_path)
    text = pytesseract.image_to_string(image)
    return text
```

####Improving our OCR

*   Removes noise
*   Improves OCR accuracy



```python
import cv2

def preprocess_image(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
    return thresh
```

```python
processed = preprocess_image("/content/ChatGPT Image Sep 3, 2025, 04_25_38 PM.png")
cv2.imwrite("processed.png", processed)

print(pytesseract.image_to_string("processed.png"))
```

####Layout via OCR Bounding Boxes

Because the above way need too much fixing and debugging

```python
import pytesseract
from pytesseract import Output
import cv2
```

```python
image = cv2.imread("/content/Corporate business invoice layout design.png")

data = pytesseract.image_to_data(image, output_type=Output.DICT)

blocks = []
n = len(data['text'])

for i in range(n):
    if int(data['conf'][i]) > 60:
        blocks.append({
            "text": data['text'][i],
            "x": data['left'][i],
            "y": data['top'][i],
            "w": data['width'][i],
            "h": data['height'][i]
        })
```

```python
data.keys()
```

```python
blocks[:10]
```

```python
blocks = sorted(blocks, key=lambda x: (x['y'], x['x']))

lines = []
current_line = []
current_y = -1

for b in blocks:
    if current_y == -1:
        current_y = b['y']

    if abs(b['y'] - current_y) < 10:
        current_line.append(b['text'])
    else:
        lines.append(" ".join(current_line))
        current_line = [b['text']]
        current_y = b['y']

if current_line:
    lines.append(" ".join(current_line))

lines[:10]
```

#####Add Basic Structure Detection

```python
structured_data = []

for line in lines:
    if line.isupper():
        structured_data.append({"type": "heading", "text": line})
    elif ":" in line:
        structured_data.append({"type": "field", "text": line})
    else:
        structured_data.append({"type": "paragraph", "text": line})

structured_data[:10]
```

##Chunking + Embeddings + FAISS

###Installation and importing

```python
!pip install sentence-transformers faiss-cpu
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
texts = [item['text'] for item in structured_data if item['text'].strip()]
texts[:5]
```

```python
embeddings = model.encode(texts)
embeddings
```

Store this data into **FAISS** (Vector DB)

```python
import faiss
import numpy as np
```

```python
embeddings.shape
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))
```

###Test Semantic Search (TSS)

```python
query = "What is the total amount?"

query_embedding = model.encode([query])

D, I = index.search(query_embedding, k=3)

for i in I[0]:
    print(texts[i])
```

### Implementing **LLM** using **API**

```python
!pip install transformers accelerate
```

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="google/flan-t5-base",
    max_length=512
)
```

```python
def answer_query(query,k=3):
  query_embedding=model.encode([query])
  D,I=index.search(query_embedding,k)

  context="\n".join([texts[i]for i in I[0]])
  prompt=f"""
  Answer the following question based on the context.
  Context:
  {context}
  Question:
  {query}
  """

  response=generator(prompt)[0]['generated_text']
  return response
```

### **Testing**

```python
print(answer_query("What is the total amount?"))
print(answer_query("What is the invoice number?"))
print(answer_query("Who is the client?"))
```
